```python
import os
import re
import json
import pandas as pd
import numpy as np
import openai
import matplotlib.pyplot as plt

#  Configuration 
openai.api_key     = os.getenv("OPENAI_API_KEY")
PAMAP_PATH         = "#1-mini.csv"
INPUT_PATTERN      = [0.48, 0.50, 0.53, 0.58, 0.62, 0.66]
TOP_N_PREFILTER    = 50   # how many windows to keep before calling GPT
STEP_PREFILTER     = 10   # slide every 10th window for speed
# 

def extract_json_array(reply: str):
    """
    Finds and parses the first JSON array in GPTs reply string.
    """
    m = re.search(r"\[.*\]", reply, flags=re.DOTALL)
    if not m:
        raise ValueError(f"No JSON array found in LLM reply:\n{reply}")
    return json.loads(m.group(0))

def load_pamap2(path=PAMAP_PATH):
    """Load PAMAP2 CSV and parse timestamps if present."""
    df = pd.read_csv(path)
    if "timestamp" in df.columns:
        df["timestamp"] = pd.to_datetime(df["timestamp"], unit="s")
    return df

def numeric_prefilter(df, pattern, top_n=TOP_N_PREFILTER, step=STEP_PREFILTER):
    """
    1) Slide windows of len(pattern) over each numeric column (subsample by step).
    2) Compute cosine similarity in NumPy.
    3) Return the top_n matches as dicts with 'column', 'start', and 'values'.
    """
    pat     = np.array(pattern)
    L       = len(pat)
    pat_norm = np.linalg.norm(pat)
    candidates = []

    for col in df.select_dtypes(include="number").columns:
        data = df[col].to_numpy()
        # dot-products via convolution
        dots  = np.convolve(data, pat[::-1], mode="valid")
        # window norms
        norms = np.sqrt(np.convolve(data**2, np.ones(L), mode="valid"))
        sims  = dots / (norms * pat_norm + 1e-12)

        # subsample windows by 'step'
        for i in range(0, sims.size, step):
            candidates.append((sims[i], col, i, data[i:i+L]))

    # pick global top_n by similarity
    candidates.sort(key=lambda x: x[0], reverse=True)
    top = candidates[:top_n]

    # format for LLM input
    return [
        {
            "column": col,
            "start": start,
            "values": window.tolist()
        }
        for _, col, start, window in top
    ]

def query_gpt_top3(input_pattern, candidates):
    """
    Send the input pattern and candidate windows to GPT4, asking for the top 3 most similar.
    """
    system = {
        "role": "system",
        "content": (
            "You are an expert at comparing numeric timeseries. "
            "Given an input pattern and a list of candidates (each with a column name, "
            "a start index, and a list of values), return the top 3 candidates most "
            "similar to the pattern. Respond ONLY with a JSON array of objects, each "
            "having keys 'column', 'start', and 'values', in descending order of similarity."
        )
    }
    user = {
        "role": "user",
        "content": (
            f"Input Pattern:\n{input_pattern}\n\n"
            f"Candidates ({len(candidates)}):\n"
            f"{json.dumps(candidates)}"
        )
    }

    resp = openai.chat.completions.create(
        model="gpt-4",
        messages=[system, user],
        temperature=0
    )
    return resp.choices[0].message.content

def plot_results(df, input_pattern, matches, timestamp_col="timestamp"):
    """
    14 figure:
       Panel0: input vs reanchored matches
       Panel13: full series of each match with window highlighted
    """
    N      = len(input_pattern)
    x_full = df[timestamp_col] if timestamp_col in df.columns else df.index

    fig, axes = plt.subplots(1, 4, figsize=(18, 4),
                             gridspec_kw={"width_ratios":[1,1,1,1]})

    # Panel0: reanchored shapes
    ax0 = axes[0]
    ax0.plot(range(N), input_pattern, label="Input", lw=3)
    for m in matches:
        ax0.plot(range(N), m["values"], "--",
                 label=f"{m['column']}@{m['start']}")
    ax0.set_title("Reanchored Shapes")
    ax0.set_xlabel("Rel. Pos")
    ax0.legend()
    ax0.grid(True)

    # Panels13: full context
    for idx, m in enumerate(matches, start=1):
        ax = axes[idx]
        col   = m["column"]
        start = m["start"]
        end   = start + N
        series = df[col]

        # full trace
        ax.plot(x_full, series, alpha=0.5)
        # highlight window
        ax.axvspan(x_full.iloc[start],
                   x_full.iloc[end-1],
                   color="orange", alpha=0.3)
        # overlay matched segment
        ax.plot(x_full.iloc[start:end],
                series.iloc[start:end],
                lw=2)
        ax.set_title(f"{col} (idx={start})")
        ax.set_xlabel(timestamp_col if timestamp_col in df.columns else "Index")
        ax.grid(True)

    fig.tight_layout()
    plt.show()

if __name__ == "__main__":
    # 1) Load data
    df = load_pamap2(PAMAP_PATH)

    # 2) Fast numeric prefilter
    candidates = numeric_prefilter(df, INPUT_PATTERN)

    # 3) LLM picks top 3
    raw_reply = query_gpt_top3(INPUT_PATTERN, candidates)
    try:
        top3 = extract_json_array(raw_reply)
    except Exception as e:
        print("Failed to parse GPT response. Raw reply:\n", raw_reply)
        raise

    # 4) Plot input vs. matches with context
    plot_results(df, INPUT_PATTERN, top3)
```

```python
import os
import re
import json
import time
import pandas as pd
import numpy as np
import ollama
import matplotlib.pyplot as plt

#  Configuration 
MODEL_NAME        = "tinyllama"
PAMAP_PATH        = "#1-mini.csv"
INPUT_PATTERN     = [0.48, 0.50, 0.53, 0.58, 0.62, 0.66]
TOP_N_PREFILTER   = 50    # how many windows to keep before LLM
STEP_PREFILTER    = 10    # slide every 10th window for speed
# 

def extract_json_array(reply: str):
    """
    Finds and parses the first JSON array in the models reply.
    """
    m = re.search(r"\[.*\]", reply, flags=re.DOTALL)
    if not m:
        raise ValueError(f"No JSON array found in LLM reply:\n{reply}")
    return json.loads(m.group(0))

def load_pamap2(path=PAMAP_PATH):
    """Load the PAMAP2 CSV and parse timestamps if present."""
    df = pd.read_csv(path)
    if "timestamp" in df.columns:
        df["timestamp"] = pd.to_datetime(df["timestamp"], unit="s")
    return df

def numeric_prefilter(df, pattern, top_n=TOP_N_PREFILTER, step=STEP_PREFILTER):
    """
    1) Slide windows of len(pattern) over each numeric column (subsample by step).
    2) Compute cosine similarity in NumPy.
    3) Return the top_n matches as dicts with 'column', 'start', and 'values'.
    """
    pat      = np.array(pattern)
    L        = len(pat)
    pat_norm = np.linalg.norm(pat)
    candidates = []

    for col in df.select_dtypes(include="number").columns:
        data = df[col].to_numpy()
        dots  = np.convolve(data, pat[::-1], mode="valid")
        norms = np.sqrt(np.convolve(data**2, np.ones(L), mode="valid"))
        sims  = dots / (norms * pat_norm + 1e-12)

        for i in range(0, sims.size, step):
            candidates.append((sims[i], col, i, data[i:i+L]))

    candidates.sort(key=lambda x: x[0], reverse=True)
    top = candidates[:top_n]

    return [
        {"column": col, "start": start, "values": window.tolist()}
        for _, col, start, window in top
    ]

def query_tinyllama_top3(input_pattern, candidates):
    """
    Send the input pattern and candidate windows to TinyLlama via Ollama,
    asking for the top 3 most similar sequences.
    Returns (reply_str, elapsed_time_s).
    """
    system_msg = (
        "You are an expert at comparing numeric timeseries. "
        "Given an input pattern and a list of candidates (each with a column name, "
        "a start index, and a list of values), return the top 3 candidates most "
        "similar to the pattern. Respond ONLY with a JSON array of objects, each "
        "having keys 'column', 'start', and 'values', sorted by descending similarity."
    )
    user_msg = (
        f"Input Pattern:\n{input_pattern}\n\n"
        f"Candidates ({len(candidates)}):\n"
        f"{json.dumps(candidates)}"
    )

    t0 = time.perf_counter()
    resp = ollama.chat(
        model=MODEL_NAME,
        messages=[
            {"role": "system", "content": system_msg},
            {"role": "user",   "content": user_msg}
        ]
    )
    elapsed = time.perf_counter() - t0
    reply   = resp["message"]["content"]
    return reply, elapsed

def plot_results(df, input_pattern, matches, timestamp_col="timestamp"):
    """
    14 figure:
       Panel0: input vs reanchored matches
       Panel13: full series of each match with window highlighted
    """
    N      = len(input_pattern)
    x_full = df[timestamp_col] if timestamp_col in df.columns else df.index

    fig, axes = plt.subplots(1, 4, figsize=(18, 4),
                             gridspec_kw={"width_ratios":[1,1,1,1]})

    # Panel0: reanchored shapes
    ax0 = axes[0]
    ax0.plot(range(N), input_pattern, label="Input", lw=3)
    for m in matches:
        ax0.plot(range(N), m["values"], "--",
                 label=f"{m['column']}@{m['start']}")
    ax0.set_title("Reanchored Shapes")
    ax0.set_xlabel("Rel. Pos")
    ax0.legend()
    ax0.grid(True)

    # Panels13: full context
    for idx, m in enumerate(matches, start=1):
        ax = axes[idx]
        col   = m["column"]
        start = m["start"]
        end   = start + N
        series = df[col]

        ax.plot(x_full, series, alpha=0.5)
        ax.axvspan(x_full.iloc[start], x_full.iloc[end-1],
                   color="orange", alpha=0.3)
        ax.plot(x_full.iloc[start:end], series.iloc[start:end], lw=2)
        ax.set_title(f"{col} (idx={start})")
        ax.set_xlabel(timestamp_col if timestamp_col in df.columns else "Index")
        ax.grid(True)

    fig.tight_layout()
    plt.show()

if __name__ == "__main__":
    start_all = time.perf_counter()

    # 1) Load data
    df = load_pamap2(PAMAP_PATH)

    # 2) Numeric prefilter
    pf_start   = time.perf_counter()
    candidates = numeric_prefilter(df, INPUT_PATTERN)
    pf_time    = time.perf_counter() - pf_start

    # 3) LLM refine via TinyLlama
    raw_reply, chat_time = query_tinyllama_top3(INPUT_PATTERN, candidates)
    try:
        top3 = extract_json_array(raw_reply)
    except Exception:
        print("Failed to parse model response. Raw reply:\n", raw_reply)
        raise

    # 4) Plot
    plot_results(df, INPUT_PATTERN, top3)

    # 5) Print timing metrics
    total_time = time.perf_counter() - start_all
    print(f"\nNumeric prefilter time: {pf_time:.2f}s")
    print(f"TinyLlama chat time:     {chat_time:.2f}s")
    print(f"Total script runtime:    {total_time:.2f}s")
```
