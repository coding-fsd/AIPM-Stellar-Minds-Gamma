```python
import pandas as pd
import ollama
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Load the cleaned data
def load_data(file_path='cleaned_data.csv'):
    """Loads the cleaned dataset."""
    return pd.read_csv(file_path)

# Step 2: Autolabel the data using TinyLlama via Ollama
def autolabel_data(df):
    """Autolabel data based on sensor readings."""
    # Convert the dataframe into a JSON format for the prompt
    data_json = df.to_json(orient='split')

    # Construct the prompt to label all sensor data features
    prompt = f"Label this dataset based on the sensor readings:\n{data_json}"

    # Call TinyLlama via Ollama (you can change to another model if needed)
    response = ollama.chat(model="tinyllama", messages=[{"role": "user", "content": prompt}])
    
    # Extract the labeled data from the response
    labeled_data = response['message']['content']
    return labeled_data

# Step 3: Plot the multivariate data
def plot_data(df):
    """Generate a graph for each sensor feature in the data."""
    # Plotting each column (sensor feature) in the dataset
    plt.figure(figsize=(10, 6))
    
    for column in df.columns:
        if column != 'timestamp':  # Exclude the timestamp column from plotting
            plt.plot(df['timestamp'], df[column], label=column)
    
    plt.xlabel('Timestamp')
    plt.ylabel('Sensor Readings')
    plt.title('Sensor Data Over Time')
    plt.legend()
    plt.grid(True)

    # Save the graph as an image
    graph_filename = 'multivariate_graph.png'
    plt.savefig(graph_filename)
    plt.close()  # Close the plot to avoid memory issues

    return graph_filename

# Step 4: Select a section of data based on user input
def select_data_section(df, start_timestamp="2022-01-01", end_timestamp="2022-01-02", start_index=None, end_index=None):
    """Select a section of data based on timestamp or index range."""
    if start_timestamp and end_timestamp:
        # Select data between two timestamps
        selected_data = df[(df['timestamp'] >= start_timestamp) & (df['timestamp'] <= end_timestamp)]
    elif start_index is not None and end_index is not None:
        # Select data based on index range
        selected_data = df.iloc[start_index:end_index]
    else:
        raise ValueError("Either timestamp or index range must be provided.")
    return selected_data

# Step 5: Calculate correlation between selected section and all features
def calculate_correlation(df, selected_section):
    """Calculate the correlation between selected section and all other features."""
    # Calculate correlation between selected section and entire dataset
    correlation = selected_section.corrwith(df.drop(columns=['timestamp']))
    return correlation

# Step 6: Plot the correlation as a heatmap
def plot_correlation(correlation):
    """Plot the correlation values as a heatmap."""
    plt.figure(figsize=(10, 6))
    sns.heatmap(correlation.to_frame(), annot=True, cmap='coolwarm', cbar=True)
    plt.title('Correlation of Selected Section with All Features')
    plt.xlabel('Feature')
    plt.ylabel('Selected Section')
    
    # Save the correlation heatmap as an image
    correlation_filename = 'correlation_heatmap.png'
    plt.savefig(correlation_filename)
    plt.close()

    return correlation_filename

# Step 7: Putting everything together
def process_data_and_generate_graph(file_path='cleaned_data.csv', start_timestamp=None, end_timestamp=None, start_index=None, end_index=None):
    """Load the data, perform autolabeling, select data section, calculate correlation, and generate the graph."""
    # Load the data
    df = load_data(file_path)
    
    # Perform autolabeling
    labeled_data = autolabel_data(df)

    # Select the data section (based on timestamp or index)
    selected_section = select_data_section(df, start_timestamp, end_timestamp, start_index, end_index)

    # Calculate correlation between the selected section and the entire dataset
    correlation = calculate_correlation(df, selected_section)

    # Plot the correlation heatmap
    correlation_filename = plot_correlation(correlation)

    # Plot the full dataset for visualization
    graph_filename = plot_data(df)

    return labeled_data, graph_filename, correlation_filename

# Step 8: Run the process and display the results
labeled_data, graph_filename, correlation_filename = process_data_and_generate_graph('cleaned_data.csv', start_timestamp="2022-01-01", end_timestamp="2022-01-02")

# Display the labeled data
print("Labeled Data:")
print(labeled_data)

# Show the graph and the correlation heatmap
from IPython.display import Image
Image(graph_filename)  # Show the full graph
Image(correlation_filename)  # Show the correlation heatmap
```

```python
import pandas as pd
import ollama
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns

# Step 1: Load the cleaned data
def load_data(file_path='cleaned_data.csv'):
    """Loads the cleaned dataset."""
    return pd.read_csv(file_path)

# Step 2: Autolabel the data using TinyLlama via Ollama
def autolabel_data(df):
    """Autolabel data based on sensor readings."""
    data_json = df.to_json(orient='split')  # Convert to JSON
    prompt = f"This dataset contains measures health monitoring data of various people. I want you to label it.  :\n{data_json}"
    response = ollama.chat(model="tinyllama", messages=[{"role": "user", "content": prompt}])
    labeled_data = response['message']['content']
    return labeled_data

# Step 3: Plot the data (For Graph Visualization)
def plot_data(df):
    """Generate a graph from the data."""
    plt.figure(figsize=(10, 6))
    plt.plot(df['timestamp'], df['heart_rate'], label='Heart Rate', color='blue')  # Customize with actual column names
    plt.xlabel('Timestamp')
    plt.ylabel('Heart Rate')
    plt.title('Heart Rate Over Time')
    plt.legend()
    plt.grid(True)

    # Save the graph as an image
    graph_filename = 'graph.png'
    plt.savefig(graph_filename)
    plt.close()

    return graph_filename

# Step 4: Select data section (Based on index or timestamp range)
def select_data_section(df, start_timestamp=None, end_timestamp=None, start_index=None, end_index=None):
    """Select a section of data based on timestamp or index range."""
    if start_timestamp and end_timestamp:
        selected_data = df[(df['timestamp'] >= start_timestamp) & (df['timestamp'] <= end_timestamp)]
    elif start_index is not None and end_index is not None:
        selected_data = df.iloc[start_index:end_index]
    else:
        selected_data = df  # If no selection criteria, use the full data
    return selected_data

# Step 5: Calculate correlations (between selected section and all features)
def calculate_correlation(df, selected_section):
    """Calculate correlation between selected section and the full dataset."""
    correlation = df.corrwith(selected_section)
    return correlation

# Step 6: Plot Correlations
def plot_correlation(correlation):
    """Plot correlation as a heatmap."""
    plt.figure(figsize=(10, 6))
    sns.heatmap(correlation.to_frame(), annot=True, cmap='coolwarm', cbar=True)
    plt.title('Correlation Heatmap')

    # Save the correlation heatmap
    correlation_filename = 'correlation_heatmap.png'
    plt.savefig(correlation_filename)
    plt.close()

    return correlation_filename

# Step 7: Putting everything together
def process_data_and_generate_graph(file_path='cleaned_data.csv', start_timestamp=None, end_timestamp=None, start_index=None, end_index=None):
    """Load the data, perform autolabeling, and generate the graph."""
    # Load the data
    df = load_data(file_path)

    # Perform autolabeling
    labeled_data = autolabel_data(df)

    # Select the data section (based on timestamp or index range)
    selected_section = select_data_section(df, start_timestamp, end_timestamp, start_index, end_index)

    # Calculate correlation between the selected section and the entire dataset
    correlation = calculate_correlation(df, selected_section)

    # Plot the correlation and save the file
    correlation_filename = plot_correlation(correlation)

    # Plot the data graph and save the file
    graph_filename = plot_data(df)

    return labeled_data, graph_filename, correlation_filename

# Step 8: Run the process and display the results
labeled_data, graph_filename, correlation_filename = process_data_and_generate_graph(
    'cleaned_data.csv', 
    start_index=80, end_index=500
)

# Display the labeled data
print("Labeled Data:")
print(labeled_data)

# Show the graph
from IPython.display import Image
Image(graph_filename)

# Show the correlation heatmap
Image(correlation_filename)
```
